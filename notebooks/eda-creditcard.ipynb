```python
import matplotlib.pyplot as plt
import pandas as pd
from src.data_loader import load_creditcard_data
import sys
sys.path.append('..')
```

```python
print("Loading creditcard.csv...")
credit_df = load_creditcard_data()
print(f"Shape: {credit_df.shape}")
print("\nFirst 5 rows:")
print(credit_df.head())
```

```python

print("\n=== Basic Information ===")
print(credit_df.info())
print("\n=== Missing Values ===")
print(credit_df.isnull().sum().sum())
```

```python
print("\n=== Class Distribution ===")
class_counts = credit_df['Class'].value_counts()
class_percent = credit_df['Class'].mean() * 100

print(f"Total transactions: {len(credit_df)}")
print(f"Legitimate (0): {class_counts[0]}")
print(f"Fraudulent (1): {class_counts[1]}")
print(f"Fraud percentage: {class_percent:.4f}%")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
class_counts.plot(kind='bar', ax=ax1, color=['green', 'red'])
ax1.set_title('Fraud vs Non-Fraud Transactions')
ax1.set_xlabel('Class')
ax1.set_ylabel('Count')

ax2.pie(class_counts, labels=['Legitimate', 'Fraud'], autopct='%1.1f%%',
        colors=['green', 'red'], startangle=90)
ax2.set_title(f'Fraud Rate: {class_percent:.4f}%')

plt.tight_layout()
plt.show()
```

```python
print("\n Transaction Amount Analysis ")
print(f"Minimum amount: ${credit_df['Amount'].min():.2f}")
print(f"Maximum amount: ${credit_df['Amount'].max():.2f}")
print(f"Average amount: ${credit_df['Amount'].mean():.2f}")
print(f"Median amount: ${credit_df['Amount'].median():.2f}")
print(
    f"\nLegitimate avg amount: ${credit_df[credit_df['Class'] == 0]['Amount'].mean():.2f}")
print(
    f"Fraud avg amount: ${credit_df[credit_df['Class'] == 1]['Amount'].mean():.2f}")
```
